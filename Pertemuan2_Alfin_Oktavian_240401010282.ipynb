```python
# Q1: Muat Dataset Titanic

import pandas as pd
import numpy as np

URL = "https://raw.github.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)
print(df.shape, df.columns.tolist())

# A1: dataset Titanic memiliki 891 baris dan 12 kolom.
```

```python
# Q2 Kolom mana yang memiliki nilai kosong (missing)?

print(df.isnull().sum())

# A2 Data set titanic memiliki 3 kolom kosong (Missing) yaitu: Age, Cabin dan Embarked
```

```python
# Q3 : Berapa persen penumpang yang selamat (Survived=1)

selamat = df[df['Survived']==1]
persentase = len(selamat)/len(df)

print ("Total Penumpang :", len(df))
print("Penumpang selamat :", df['Survived'].sum())
print("Persentase selamat (%) :", persentase * 100)

# A3 Presentase penumpang yang berhasil selamat yaitu 38.38%
```

```python
# Q4: Rata-rata usia penumpang wanita kelas 1 yang selamat

wanita_kelas1_selamat = df[
    (df['Sex'] == 'female') &
    (df['Pclass'] == 1) &
    (df['Survived'] == 1)
]
({
    "Jumlah Penumpang": [len(wanita_kelas1_selamat)],
    "Rata-rata Usia": [wanita_kelas1_selamat['Age'].mean()]
})

# A4 Rata-rata usia wanita yang selamat adalah 35 tahun
```

```python
# Q5: Kelas penumpang mana yang tingkat keselamatan tertinggi ?

hasil = df.groupby('Pclass').agg(
        total=('PassengerId', 'count'),
        selamat=('Survived','sum'),
        persen = ('Survived','mean'),
).round(3)
print(hasil)

# A5 tingkat keselamatan tertinggi berada di kelas 1 total 136 yang selamat
```
